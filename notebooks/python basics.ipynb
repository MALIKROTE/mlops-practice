```python
# Dictionaries
person = {
    "name": "Malik",
    "age": 30,
    "city": "Pune"
}
print(person["name"])
print(person["age"])
print(person["city"])
```

```python
# Loop + DIctionaries
student = {
    "name": "Malik",
    "age": 30,
    "city": "Pune"
}
for key, value in student.items():
    print(key, ":", value)
```
