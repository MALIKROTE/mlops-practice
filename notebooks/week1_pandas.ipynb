```python
# Features and targets
import pandas as pd

df = pd.DataFrame({
    "age":[25,30,35,40],
    "experience":[2,5,8,10],
    "salary":[30000,50000,70000,90000]
})

x=df[["age","experience"]]
y=df[["salary"]]

print(x)
print(y)
```

```python
# Train/Test Split
# to practice this we need to install scikit-learn using pip install scikit-learn

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.25,
    random_state=42
)

print(y_train)
print(x_test)

```
